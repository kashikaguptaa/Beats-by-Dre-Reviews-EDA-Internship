## 6. Sentiment Analysis

### 6.1 Methodology
Explain the sentiment analysis techniques used (e.g., TextBlob, NLTK) and provide code snippets.


I used TextBlob for sentiment analysis in this project. TextBlob is a Python library that makes it easy to perform common natural language processing (NLP) tasks, including sentiment analysis. It evaluates the polarity of a given piece of text, which ranges from -1 (very negative) to 1 (very positive). This makes it useful for determining the overall sentiment expressed in customer reviews.

```python
from textblob import TextBlob
# Example DataFrame (replace with your actual data)
# df = pd.read_csv('your_dataset.csv')

# Assuming 'content' column contains the text reviews
# Function to classify sentiment based on TextBlob polarity
def classify_sentiment(text):
    analysis = TextBlob(text)
    # Get the polarity score
    polarity = analysis.sentiment.polarity
    # Classify as positive, negative, or neutral
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the sentiment analysis to the 'content' column
df['sentiment'] = df['content'].apply(classify_sentiment)
```

## 7. AI-Generated Insights

### 7.1 Using Gemini AI
Briefly explain how Gemini AI was used to extract insights and include any relevant code snippets or explanations.

I utilized Gemini AI to extract detailed insights from the customer reviews by first setting up an API key through Google AI Studio. Once the API key was configured, I loaded my dataset and implemented a function that allowed me to send the review text to the Gemini AI API. The API then analyzed the content and returned insights regarding consumer sentiment, key themes, and other relevant data points. This process streamlined the analysis, providing a deeper understanding of customer feedback for Beats by Dre products.

```python
import pandas as pd
import google.generativeai as genai

# Load your dataset
df = pd.read_csv('/content/merged_rawfile1')

# Replace 'your_api_key' with the actual API key
api_key = "Api key"

# Function to analyze reviews using Gemini AI
def analyze_reviews(reviews, api_key,prompt_message):
    genai.configure(api_key=api_key)

    model = genai.GenerativeModel(
        model_name="gemini-1.5-flash",
        generation_config={
            "temperature": 1,
            "top_p": 0.95,
            "top_k": 64,
            "max_output_tokens": 8192,
            "response_mime_type": "text/plain",
        },
    )

    chat_session = model.start_chat(history=[])

    prompt = prompt_message + "\n\n".join(reviews)
    response = chat_session.send_message(prompt)
    return response.text
```

```python
from textblob import TextBlob
# Example DataFrame (replace with your actual data)
# df = pd.read_csv('your_dataset.csv')

# Assuming 'content' column contains the text reviews
# Function to classify sentiment based on TextBlob polarity
def classify_sentiment(text):
    analysis = TextBlob(text)
    # Get the polarity score
    polarity = analysis.sentiment.polarity
    # Classify as positive, negative, or neutral
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the sentiment analysis to the 'content' column
df['sentiment'] = df['content'].apply(classify_sentiment)
```

```python
import pandas as pd



# Group by brand and sentiment
grouped_sentiment = df.groupby(['product_name', 'sentiment']).size().unstack(fill_value=0)

# Display the grouped sentiment counts
print(grouped_sentiment)
```

```python
# Calculate the proportion of each sentiment for each brand
sentiment_distribution = grouped_sentiment.div(grouped_sentiment.sum(axis=1), axis=0) * 100

# Display the sentiment distribution
print(sentiment_distribution)
```
